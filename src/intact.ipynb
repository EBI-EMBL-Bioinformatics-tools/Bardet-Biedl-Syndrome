```python
# Libs

import pandas
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
data_source = "../data/intact/"
data_filename = "bbs6-bbs10-bbs12-2024-11-08-20-14.txt"

data = pandas.read_csv(data_source + data_filename, sep="\t", header=0)

data.head()
```

### <b>Columns removed from data</b>

We assume that both interactors B and A are from Taxon id 9606 (Homo sapiens) as this filter was applied when the search of data was performed, therefore we will exclude these columns "Taxid interactor A", "Taxid interactor B".

We also know the search was performed with the filter of interactor (A or B) as either of BBS6, BBS10 or BBS12, therefore we can remove the following columns: "Alt ID(s) interactor A", "Alt ID(s) interactor B", "Alias(es) interactor A", "Alias(es) interactor B".

```python
# Columns to remove: "Taxid interactor A", "Taxid interactor B", "Alt ID(s) interactor A", "Alt ID(s) interactor B", "Alias(es) interactor A", "Alias(es) interactor B"

features_to_remove = ["Taxid interactor A", "Taxid interactor B", "Alt. ID(s) interactor A", "Alt. ID(s) interactor B", "Alias(es) interactor A", "Alias(es) interactor B", "Publication 1st author(s)"]

can_remove = True
for feature in features_to_remove:
    if feature not in data.columns:
        can_remove = False
        break

if can_remove:
    data = data.drop(columns=features_to_remove)
    
data.head()
```

Now we replace the ids of the interactor with the name of the gene instead to facilite reading

- BBS6/MKKS: Q9NPJ1
- BBS10: Q8TAM1   
- BBS12: Q6ZW61

```python
# BBS6/MKKS: Q9NPJ1, BBS10: Q8TAM1, BBS12: Q6ZW61

bbs6_id = "Q9NPJ1"
bbs10_id = "Q8TAM1"
bbs12_id = "Q6ZW61"

# Replace "ID(s) interactor A" and "ID(s) interactor B" by the gene name

# Change column name from "ID(s) interactor A" to "Gene name"
data = data.rename(columns={"ID(s) interactor A": "Gene name A", "ID(s) interactor B": "Gene name B"})

# Example start value: "uniprotkb:Q8TAM1"
data["Gene name A"] = data["Gene name A"].str.replace("uniprotkb:", "")
data["Gene name B"] = data["Gene name B"].str.replace("uniprotkb:", "")

# Remplace ids by gene names
data["Gene name A"] = data["Gene name A"].replace(bbs6_id, "BBS6")
data["Gene name B"] = data["Gene name B"].replace(bbs6_id, "BBS6")
data["Gene name A"] = data["Gene name A"].replace(bbs10_id, "BBS10")
data["Gene name B"] = data["Gene name B"].replace(bbs10_id, "BBS10")
data["Gene name A"] = data["Gene name A"].replace(bbs12_id, "BBS12")
data["Gene name B"] = data["Gene name B"].replace(bbs12_id, "BBS12")

data.head()
```

### <b>Filter data tuples</b>

Remove tuples with a miscore < 0.4

```python
# Remove tuples with values < 0.4

"""
Example values of "Confidence value(s)" column:
- "intact-miscore:0.56|author-based confidence:0.868278297126"
- "intact-miscore:0.56"

Not all rows have the "author-based confidence" value
"""

print(f"Number of rows before filtering >0.4 miscore: {len(data)}")

# 1. Create column "IntAct Miscore" 
data["IntAct Miscore"] = data["Confidence value(s)"].str.extract(r"intact-miscore:(\d\.\d+)")
data["IntAct Miscore"] = data["IntAct Miscore"].astype(float)

# 2. Create column "Autor-based confidence"
data["Autor-based confidence"] = data["Confidence value(s)"].str.extract(r"author-based confidence:(\d\.\d+)")
data["Autor-based confidence"] = data["Autor-based confidence"].astype(float)

# 3. Remove "Confidence value(s)" column
data = data.drop(columns=["Confidence value(s)"])

# 4. Remove tuples with values < 0.4
data = data[data["IntAct Miscore"] >= 0.4]

print(f"Number of rows after filtering >0.4 miscore: {len(data)}")
```
